```python
# ============================================================
# CREDIT CARD FRAUD DETECTION
# COMPLETE EDA + DATA PREPROCESSING
# (CORRECTED VERSION)
# ============================================================

# =========================
# 1. IMPORT LIBRARIES
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import (
    MinMaxScaler,
    StandardScaler,
    LabelEncoder
)

from scipy.stats import zscore


# ============================================================
# 2. LOAD DATASET
# ============================================================
df = pd.read_csv("RAW_Dataset_With_Account_Number.csv")

print("\nDataset Loaded Successfully")
print("Dataset Shape:", df.shape)


# ============================================================
# 3. BASIC EDA
# ============================================================
print("\nFirst 5 Rows")
print(df.head())

print("\nDataset Information")
print(df.info())

print("\nStatistical Summary")
print(df.describe(include='all'))


# ============================================================
# 4. CHECK MISSING VALUES
# ============================================================
print("\nColumns with Missing Values")

null_cols = df.columns[df.isnull().sum() > 0]

print(df[null_cols].isnull().sum())


# ============================================================
# 5. VISUALIZE MISSING VALUES
# ============================================================
plt.figure(figsize=(12,6))

sns.heatmap(df.isnull(), cbar=False)

plt.title("Missing Values Heatmap")

plt.show()


# ============================================================
# 6. HANDLE MISSING VALUES
# ============================================================
for col in df.columns:

    # Numerical columns
    if df[col].dtype in ['int64', 'float64']:

        df[col] = df[col].fillna(df[col].mean())

    # Categorical columns
    else:

        df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing Values Replaced Successfully")


# ============================================================
# 7. HANDLE DUPLICATES
# ============================================================

# Remove duplicates only from Transaction_ID

if 'Transaction_ID' in df.columns:

    df = df.drop_duplicates(subset=['Transaction_ID'])

print("\nDuplicate Transaction_ID Removed")
print("New Shape:", df.shape)

# Customer_ID and Account_Number duplicates retained


# ============================================================
# 8. HANDLE CATEGORICAL COLUMNS
# ============================================================

# Select categorical columns
cat_cols = df.select_dtypes(include='object').columns

# Exclude ID columns from encoding
exclude_encoding = [
    'Transaction_ID',
    'Customer_ID',
    'Account_Number'
]

cat_cols = [
    col for col in cat_cols
    if col not in exclude_encoding
]

print("\nCategorical Columns To Encode:")
print(cat_cols)

# Label Encoding
le = LabelEncoder()

for col in cat_cols:

    df[col] = le.fit_transform(df[col])

print("\nCategorical Columns Encoded Successfully")


# ============================================================
# 9. DETECT BINARY COLUMNS
# ============================================================
binary_cols = []

for col in df.columns:

    unique_vals = df[col].dropna().unique()

    if len(unique_vals) == 2:

        binary_cols.append(col)

print("\nBinary Columns:")
print(binary_cols)


# ============================================================
# 10. EXCLUDE ID + BINARY COLUMNS
# ============================================================
exclude_cols = [
    'Transaction_ID',
    'Customer_ID',
    'Account_Number'
]

# Continuous numeric columns only
num_cols = [
    col for col in df.select_dtypes(include=np.number).columns
    if col not in exclude_cols
    and col not in binary_cols
]

print("\nContinuous Numerical Columns:")
print(num_cols)


# ============================================================
#               GRAPHS BEFORE EDA
# ============================================================

# ============================================================
# 11. HISTOGRAM BEFORE EDA
# ============================================================
df[num_cols].hist(figsize=(14,10))

plt.suptitle("Histogram Before EDA")

plt.show()


# ============================================================
# 12. BOXPLOT BEFORE EDA
# ============================================================
plt.figure(figsize=(15,6))

sns.boxplot(data=df[num_cols])

plt.title("Boxplot Before EDA")

plt.xticks(rotation=90)

plt.show()


# ============================================================
# 13. CORRELATION HEATMAP BEFORE EDA
# ============================================================
plt.figure(figsize=(14,10))

sns.heatmap(
    df[num_cols].corr(),
    annot=False,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap Before EDA")

plt.show()


# ============================================================
# 14. PIE CHART FOR BINARY COLUMN
# ============================================================
if 'Is_International' in df.columns:

    plt.figure(figsize=(7,7))

    df['Is_International'].value_counts().plot.pie(
        autopct='%1.1f%%'
    )

    plt.title("International vs Domestic Transactions")

    plt.ylabel("")

    plt.show()


# ============================================================
#               OUTLIER DETECTION USING Z-SCORE
# ============================================================

# ============================================================
# 15. Z-SCORE CALCULATION
# ============================================================
z_scores = np.abs(zscore(df[num_cols]))

z_df = pd.DataFrame(z_scores, columns=num_cols)

print("\nZ-Score Values")
print(z_df.head())


# ============================================================
# 16. Z-SCORE GRAPH
# ============================================================
for col in num_cols:

    plt.figure(figsize=(10,5))

    plt.plot(z_df[col])

    plt.axhline(
        y=3,
        color='r',
        linestyle='--'
    )

    plt.title(f"Z-Score Graph for {col}")

    plt.xlabel("Rows")

    plt.ylabel("Z-Score")

    plt.show()


# ============================================================
# 17. REMOVE OUTLIERS
# ============================================================

mask = (z_scores < 3).all(axis=1)

df = df[mask]

print("\nOutliers Removed Using Z-Score")

print("Shape After Removing Outliers:", df.shape)


# ============================================================
#               NORMALIZATION
# ============================================================

# ============================================================
# 18. NORMALIZATION
# ============================================================
scaler = MinMaxScaler()

normalized_values = scaler.fit_transform(df[num_cols])

df_normalized = pd.DataFrame(
    normalized_values,
    columns=[col + "_Normalized" for col in num_cols]
)

print("\nNormalized Values")

print(df_normalized.head())


# ============================================================
# 19. NORMALIZATION GRAPH
# ============================================================
plt.figure(figsize=(10,5))

df_normalized.head(50).plot()

plt.title("Normalized Values Graph")

plt.xlabel("Rows")

plt.ylabel("Normalized Values")

plt.show()


# ============================================================
#               STANDARDIZATION
# ============================================================

# ============================================================
# 20. STANDARDIZATION
# ============================================================
std_scaler = StandardScaler()

standardized_values = std_scaler.fit_transform(df[num_cols])

df_standardized = pd.DataFrame(
    standardized_values,
    columns=[col + "_Standardized" for col in num_cols]
)

print("\nStandardized Values")

print(df_standardized.head())


# ============================================================
# 21. STANDARDIZATION GRAPH
# ============================================================
plt.figure(figsize=(10,5))

df_standardized.head(50).plot()

plt.title("Standardized Values Graph")

plt.xlabel("Rows")

plt.ylabel("Standardized Values")

plt.show()


# ============================================================
#               AFTER EDA GRAPHS
# ============================================================

# ============================================================
# 22. HISTOGRAM AFTER EDA
# ============================================================
df_standardized.hist(figsize=(14,10))

plt.suptitle("Histogram After EDA")

plt.show()


# ============================================================
# 23. BOXPLOT AFTER EDA
# ============================================================
plt.figure(figsize=(15,6))

sns.boxplot(data=df_standardized)

plt.title("Boxplot After EDA")

plt.xticks(rotation=90)

plt.show()


# ============================================================
# 24. LINE GRAPH
# ============================================================
plt.figure(figsize=(10,5))

df_standardized.head(50).plot()

plt.title("Line Graph After EDA")

plt.show()


# ============================================================
# 25. CORRELATION HEATMAP AFTER EDA
# ============================================================
plt.figure(figsize=(14,10))

sns.heatmap(
    df_standardized.corr(),
    annot=False,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap After EDA")

plt.show()


# ============================================================
# 26. FINAL DATASET
# ============================================================

# Combine:
# 1. Cleaned original dataset
# 2. Normalized dataset
# 3. Standardized dataset

final_df = pd.concat(
    [
        df.reset_index(drop=True),
        df_normalized.reset_index(drop=True),
        df_standardized.reset_index(drop=True)
    ],
    axis=1
)


# ============================================================
# 27. FINAL SAFETY CHECK
# ============================================================
print("\nRemaining Missing Values:")

print(final_df.isnull().sum().sum())

# Remove accidental NaNs if any
final_df = final_df.dropna()


# ============================================================
# 28. SAVE FINAL DATASET
# ============================================================
final_df.to_csv(
    "Cleaned_Normalized_Standardized_Dataset.csv",
    index=False
)

print("\nFinal Cleaned Dataset Saved Successfully")

print("File Name: Cleaned_Normalized_Standardized_Dataset.csv")


# ============================================================
# 29. FINAL DATASET INFORMATION
# ============================================================
print("\nFinal Dataset Shape:")
print(final_df.shape)

print("\nFinal Dataset Preview:")
print(final_df.head())
```
